```python
from pathlib import Path
import json
path = Path("./BARDQAtempDataset.json")

# Open .json file
with open(path, 'rb') as f:
    bardqa_dict = json.load(f)
```

```python
from random import sample

sample_data = sample(bardqa_dict['data'],100)
for data in sample_data:
    print("-----------------------------------------------------------")
    print("CONTEXT ----> ",data['sent_with_ans'])
    print("QUESTION ----> ",data['questions'][0])
    print("ANSWER ----> ", data['answer']['text'])
    print("-----------------------------------------------------------")
```
