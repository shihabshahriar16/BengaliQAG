# Consider all categories in directory

```python
context_data = {}
context_data['data'] = []
```

```python
import os
# assign directory
directory = './entertainment'
texts = []
 
# iterate over files in
# that directory
for filename in os.listdir(directory):
    txt_file = os.path.join(directory, filename)
    with open(txt_file) as f:
        lines = f.readlines() # list containing lines of file
    lines = [line.strip() for line in lines]
    texts.append(" ".join(lines).strip())
```

```python
texts2 = []
for txt in texts:
    txt = txt.replace("।বিজ্ঞপ্তি ।", "।")
    txt = txt.replace("। বিজ্ঞপ্তি।", "।")
    txt = txt.replace("।  বিজ্ঞপ্তি।", "।")
    txt = txt.replace("। বিজ্ঞপ্তি ।", "।")
    txt = txt.replace("।বিজ্ঞপ্তি।", "।")
    txt = txt.replace("।বিজ্ঞপ্তি", "।")
    txt = txt.replace("। বিজ্ঞপ্তি", "।")
    txt = txt.replace("।’ বিজ্ঞপ্তি", "।")
    txt = txt.replace("।’বিজ্ঞপ্তি", "।")
    txt = txt.replace("। ’ বিজ্ঞপ্তি", "।")
    txt = txt.replace("-বিজ্ঞপ্তি", "")
    txt = txt.replace("–বিজ্ঞপ্তি", "")
    txt = txt.replace("। প্রেস বিজ্ঞপ্তি", "।")
    txt = txt.replace("। খবর বিজ্ঞপ্তি ।", "।")
    txt = txt.replace("। খবর বিজ্ঞপ্তি।", "।")
    txt = txt.replace("বিজ্ঞপ্তি", "")
    txt = txt.strip()
    texts2.append(txt)
```

```python
import random
texts2 = [" ".join(data.split()) for data in texts2]
texts2 = [data for data in texts2 if 50<=len(data)<1500]
texts2 = random.sample(texts2, 4000)
for txt in texts2:
    data = {}
    data['title'] = directory.split("/")[-1]
    data['context'] = txt
    context_data['data'].append(data)
```
